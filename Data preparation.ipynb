```python
import sqlite3 as lite
import sys
import pandas as pd
import numpy as np
import datetime
import json
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
# Select what symbol to analyze for
symbol = "DNA"

con = None
con = lite.connect('/hdd/db/crypto.db') #Connect to DB


with con: 
    df = pd.read_sql_query('SELECT * FROM {}_tweets'.format(symbol), con=con, index_col='time')
    
df = df.drop(['index'], axis=1)
```

```python
# Addition of crypto-specific terms to lexicon
crypto_terms = {
    'bullish': 3.0,
    'bull': 2.0,
    'bearish': -3.0,
    'bear': -2.0,
    'hodl': 2.0,
    'lambo': 3.0,
    'moon': 3.0,
    'mooning': 3.0,
    'btd': 3.0,
    'dump': -2.0,
    'dumping': -2.0,
    'pump': 2.0,
    'pumping': 2.0,
    'p&d': -1.0,
    'fomo': 3.0,
    'fud': -3.0,
    'fudster': -3.0,
    'skyrocket': 3.0,
    'shilling': 2.5,
    'bagholder': -1.0,
    'dca': 1.5,
    'shitcoin': -3.0    
}


# Calculate sentiment using VADER (lexical)
analyzer = SentimentIntensityAnalyzer()
analyzer.lexicon.update(crypto_terms)

total_compound = 0
total_neu = 0 
total_neg = 0
total_pos = 0
Sent_Neu, Sent_Neg, Sent_Pos, Sent_Compound = ([] for i in range(4))

for tweet in df['Tweets']:
    score = analyzer.polarity_scores(tweet)
    Sent_Neu.append(score['neu'])
    Sent_Neg.append(score['neg'])
    Sent_Pos.append(score['pos'])
    Sent_Compound.append(score['compound'])

df['Sent_Neu'] = Sent_Neu
df['Sent_Neg'] = Sent_Neg
df['Sent_Pos'] = Sent_Pos
df['Sent_Compound'] = Sent_Compound
```
