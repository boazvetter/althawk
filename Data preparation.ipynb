```python
import sqlite3 as lite
import sys
import pandas as pd
import numpy as np
import datetime
import json
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
con = None
con = lite.connect('/hdd/db/crypto.db') #Connect to DB


with con: 
    df = pd.read_sql_query('SELECT * FROM ETH_tweets', con=con, index_col='time')
    
df = df.drop(['index'], axis=1)
```

```python
df
```

```python
analyzer = SentimentIntensityAnalyzer()
total_compound = 0
total_neu = 0 
total_neg = 0
total_pos = 0
for tweet in df['Tweets']:
    score = analyzer.polarity_scores(tweet)
    total_compound = total_compound + score['compound']
    total_neu = total_neu + score['neu']
    total_neg = total_neg + score['neg']
    total_pos = total_pos + score['pos']
    #print("{:-<65} {}".format(tweet, str(score)))
    
print("avg Neu: ", total_neu / len(df['Tweets']))
print("avg Neg: ", total_neg / len(df['Tweets']))
print("avg Pos: ", total_pos / len(df['Tweets']))
print("avg Compound: ", total_compound / len(df['Tweets']))
```
